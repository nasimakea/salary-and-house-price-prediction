```python



import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/py/ML/2_linear_reg_multivariate/homeprices.csv")
```

```python
df
```

```python
df.fillna(np.mean(df['bedrooms']),inplace=True)
```

```python
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df[['price']])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
reg.predict([[1500,4,5]])
```

```python
df1=pd.read_csv("/content/py/ML/2_linear_reg_multivariate/Exercise/hiring.csv")
```

```python
df1
```

```python
df1["test_score(out of 10)"].fillna(np.mean(df1["test_score(out of 10)"]),inplace=True)
```

```python
df1
```

```python
word_to_number = {
'five':5,'two':2,'seven':7,'three':3,'ten':10,'eleven':11

                   }
```

```python
df1 = df1.replace(word_to_number)

print(df1)
```

```python
df1.fillna(np.mean(df1['experience']),inplace=True)
```

```python
df1
```

```python
reg=linear_model.LinearRegression()
reg.fit(df1[['experience','test_score(out of 10)','interview_score(out of 10)']],df1['salary($)'])
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
